## FETCH WORD DOCUMENT CONTENTS

```python
import docx
```

```python
def obtainText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```
